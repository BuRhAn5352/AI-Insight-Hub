CHECKING TREND

```python
df['Order Year']=df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['YearMonth'] = df['Order Date'].dt.to_period('M')
```

```python
df
```

```python
monthly_sales = df.groupby('YearMonth')[['Sales', 'Profit']].sum().reset_index()
monthly_sales
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['YearMonth'].astype(str) , monthly_sales['Sales'] ,label ='Sales')  # cant be period if u wanna plot so we did conversion to str 
plt.plot(monthly_sales['YearMonth'].astype(str) , monthly_sales['Profit'] , label = 'Profit')
plt.xticks(rotation=45)
plt.title('Monthly Sales and Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
cat_trend = df.groupby(['YearMonth', 'Category'])['Sales'].sum().reset_index()
cat_trend
```

```python
cat_trend['YearMonth'] = cat_trend['YearMonth'].astype(str)  # Convert Period to string

plt.figure(figsize=(12, 6))
sns.lineplot(x='YearMonth', y='Sales', hue='Category', data=cat_trend)
plt.xticks(rotation=45)
plt.title('Monthly Sales Trend by Category')
plt.show()
```

```python
region_trend = df.groupby(['YearMonth', 'Region'])['Profit'].sum().reset_index()
region_trend['YearMonth']=region_trend['YearMonth'].astype(str)
plt.figure(figsize=(12,6))
sns.lineplot(x='YearMonth', y='Profit', hue='Region', data=region_trend)
plt.xticks(rotation=45)
plt.tight_layout()
plt.title('Monthly Profit Trend by Region')
plt.show()
```

What is rolling(3)?
In pandas, .rolling(3) creates a rolling window of size 3 over your data.
It means: for each value, look at that value and the previous 2 values (so, 3 in total), and then apply a calculation (like mean(), sum(), etc.).

Example: If your sales are [10][20][30][40][50], the 3-point rolling mean would be:

First two values: NaN (not enough data)

Third value: mean of [10, 20, 30] = 20

Fourth value: mean of [20, 30, 40] = 30

Fifth value: mean of [30, 40, 50] = 40

This is called a moving average and helps smooth out short-term fluctuations to show the overall trend.  

Rolling window: To smooth noisy data and see trends more clearly.
Alpha: To make plots easier to read when lines overlap

```python
monthly_sales['Sales_MA3'] = monthly_sales['Sales'].rolling(3).mean()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['YearMonth'].astype(str), monthly_sales['Sales'], alpha=0.4, label='Actual Sales')
plt.plot(monthly_sales['YearMonth'].astype(str), monthly_sales['Sales_MA3'], label='3-Month Moving Avg', linewidth=2)
plt.legend()
plt.tight_layout()
plt.title('Smoothed Sales Trend')
plt.xticks(rotation=45)
plt.show()
```

```python
#checking skewness of the sale to check how is it
plt.hist(df['Sales'], bins=50)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Sales'].plot(kind='kde')
plt.title('Sales Density Plot')
plt.show()
```

```python
skewness = df['Sales'].skew()
print(f'Skewness before log transformation: {skewness}')
```
